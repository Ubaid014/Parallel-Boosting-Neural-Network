## Hyperparameter tunning for XgBoost algorithm

```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
import xgboost as xgb
import gc
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("Islamabad_featured_data.csv")
df.head()
```

```python
df.dropna(inplace=True)
```

```python
y= df[['All sky irradiance']]
```

```python
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(df)

scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1)).flatten() 
```

```python
X_train,X_test,y1_train,y1_test = train_test_split(X_scaled,y_scaled,test_size=0.2,shuffle=False)
```

```python
xgb= xgb.XGBRegressor() 
```

```python
n_estimators= [100,200,300]
learning_rate= [0.01,0.03,0.1,0.5]
max_depth=[6,9,12]
subsamples = [0.7,0.8,0.9,1]
colsample_bytree=[0.6,0.8,0.9,1]
tree_method= ['gpu_hist','approx']
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'subsamples': subsamples,
              'colsample_bytree':colsample_bytree,
              'learning_rate':learning_rate,
              'tree_method':tree_method
             }
print(param_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV

xgb_grid = RandomizedSearchCV(estimator = xgb, 
                       param_distributions = param_grid, 
                       cv = 3, 
                       verbose=2, 
                       n_jobs = -1)
```

```python
xgb_grid.fit(X_train,y1_train)
```

```python
xgb_grid.best_params_
```

## Catboost Hyperparameter

```python
import pandas as pd
import numpy as np
import os
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
import gc
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("Islamabad_featured_data.csv")
df.head()
```

```python
df.dropna(inplace=True)
```

```python
y= df[["All sky irradiance"]]
```

```python
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(df)

scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1)).flatten() 
```

```python
X_train,X_test,y1_train,y1_test = train_test_split(X_scaled,y_scaled,test_size=0.2,shuffle= False)
```

```python
CatBoost= CatBoostRegressor()
```

```python
n_estimators= [1000,1250,1500]
learning_rate= [0.1,0.3,0.01,0.05]
max_depth=[2,4.6,8]
l2_leaf_reg=[0.2,0.5,1,3]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
            
              'learning_rate':learning_rate,
              'l2_leaf_reg':l2_leaf_reg
             }
print(param_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV

CatBoost_grid = RandomizedSearchCV(estimator = CatBoost, 
                       param_distributions = param_grid, 
                       cv = 3, 
                       verbose=2, 
                       n_jobs = -1)
```

```python
CatBoost_grid.fit(X_train,y1_train)
```

```python
CatBoost_grid.best_params_
```
